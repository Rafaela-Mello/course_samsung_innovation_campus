**Comparação com a expressão de dados no dicionário**

**Definições: Lista, índice, indexação**

Lista
- Estrutura de dados que pode conter um valor individual em uma única variável ou valor de dados.
- Suporta a extração dos valores necessários dos itens salvos e substituir valores internos.
- Os itens ou elementos formam uma lista.
- Os valores (ou itens) são distinguidos por meio de uma vírgula dentro do [ ].
- É possível importar valores desejados utilizando um índice para designar a localização do primeiro valor.

Índice
- Refere-se ao número que indica o valor do item da lista.
- O índice de uma lista com n itens aumentará de 0 para n-1.
- O uso de um índice negativo é possível.

Indexação

## **Unidade 11: Dicionários**

**Definição de Dicionário**
- É um tipo de dado que possui pares de valor-chave.
- As chaves não podem ser duplicadas.
- O item de dicionário é definido por estar conectado com dois pontos como [chave] : [valor].
- Consiste em mais de um item.
- O [valor] pode ser pesquisado usando a [chave], mas o [valor] não pode ser usado para pesquisar a [chave].

**Comparação da lista e do dicionário**

```python
# Sintaxe dicionário: nome do dicionário = {[altura  1] : [valor 1],  ...}
pessoa = {'Nome': 'David Doe', 'Idade': 26, 'Peso': 82}
pessoa
```

**Funções e operadores para o dicionário**
- **Função len**: Utilizado para obter o número de itens no dicionário
- **Operador in**: Retorna True se a chave estiver presente no dicionário
- **Operador not in**: Retorna True se a chave não for encontrada no dicionário
- **Operadores ==, !=**: Verifica se dois dicionários têm os mesmos itens
- **Comando clear**: Apagar todos os itens

```python
# Função len
pessoa = {'Nome': 'David Doe', 'Idade': 26, 'Peso': 82}
len(pessoa)
```

```python
# Função in
pessoa = {'Nome': 'David Doe', 'Idade': 26, 'Peso': 82}
print('Nome' in pessoa) #'Nome' é encontrado como chave
print("Trabalho"in pessoa) #'Trabalho' não é encontrado como chave
print('David Doe' in pessoa) #'David Doe' não é chave é valor!
```

```python
# Função not in
pessoa = {'Nome': 'David Doe', 'Idade': 26, 'Peso': 82}
print('Nome' not in pessoa) #'Nome' é encontrado como chave
print("Trabalho" not in pessoa) #'Trabalho' não é encontrado como chave
print('David Doe' not in pessoa) #'David Doe' não é chave é valor!
```

```python
# Operador == e operador !=
d1 = {'Nome': 'David Doe', 'Idade' : 26}
d2 = {'Idade' : 26, 'Nome': 'David Doe'}
print(d1 == d2)
print(d1!=d2)
```

```python
# Comando clear
dic = {0:11, 1:22, 2:33, 3:44, 4:55}
print(dic)
dic.clear()
print(dic)
```

**Dicionário de dados json**
- JSON (JavaScript Object Notation) é um formato de texto em que se pode ler e escrever os dados mais rapidamente do que XML
- Foi desenvolvido para fornecer um objeto de dados que consiste em "pares de valor-chave".

- Ao usar o módulo json em um código, use o comando "import json".
- Os dados do tipo json consistem em strings que podem ser convertidos em tipo dicionário através da função de loads no módulo json.
- Para ler strings do tipo json, é utilizada a função de loads.
- Use a função dump do módulo json para escrever dados de dicionário como arquivo tipo json.

```python
# Tentando usar dados em json
import json
data ='{"Name":"David Doe","Age":25,"Hobby":["Basketball"],\
"Family": {"Father":"John Doe","Mother":"Marry Doe"},"Married": true}'

json_data = json.loads(data)
print(type(json_data))
print(json_data['Name'])
print(json_data['Family'])
print(json_data['Married'])
```

```python
# Converter dados json em dicionário
import json
data =  '{"Título": "O Velho e o Mar", "ISBN": "12345", "Autor": "Ernest Hemingway"}'
json_data = json.loads(data)

# Use with e a função open para criar os dados de texto de book.json no modo de escrita ('w').
# Use o comando json.dump para gravar os dados de json_data no arquivo "f".
# Código para criar json_ data como arquivo book.json
with open('book.json', 'w') as f:
  json.dump(json_data, f, indent='\t')
```

- Para a troca, é usada uma variável temporária.

```python
a = 100
b = 200
print('antes da troca : a = ', a, 'b = ', b)

temp = a
a = b
b = temp
print('after swap : a =',  a, 'b = ', b)
```

  - Os dados são geralmente armazenados em buffer e depois processados no processador responsável pela entrada/saída do computador.
  - O buffer utilizado aqui é um espaço de memória temporário.
  - Se um arquivo for carregado no buffer, que é um espaço de memória temporário, a memória do sistema continuará a ser utilizada.
  - Para escrever arquivos com eficiência, ele é usado para coletar dados até um certo tamanho e lê-los ou escrevê-los todos de uma só vez.
  - Se o conteúdo do buffer for enviado para o disco através do método close, tanto a memória quanto o tamanho do buffer são esvaziados.
  - Portanto, a eficiência da memória aumenta somente quando o arquivo é fechado usando o método close.

## **Unidade 14: Métodos de Dicionários 1**

**Adicionar, editar tipo de dados do dicionário**
- É possível acrescentar e apagar informações no dicionário, conforme necessário.
- Uma das funções importantes para fazer isso é adicionar um par de valores-chave.
- Métodos avançados de dicionário:
  - setdefault: adiciona pares de valores chave padrão
  - update: modifica o valor de uma chave e adiciona um par de valores de chave se não houver chave.

**Método setdefault**

```python
# Exemplo: conta o número de letras alfabéticas na lista str1 dada e as armazena.

# Código do dicionário sem setdefault.
str1 = ['a', 'b', 'c', 'b', 'a', 'b', 'c']
dic = {}
for ch in str1:
   if ch not in dic.keys():
     dic[ch] = 0
   dic[ch] +=1
print ('alphabet counting :', dic)
```

- Evita cláusulas condicionais de 'if' do código anterior.
- Se o valor chave é passado como primeiro parâmetro e o valor padrão como segundo parâmetro como segue, o valor padrão para a chave é especificado como 0.

```python
# Código de dicionário com setdefault
str1 = ['a', 'b', 'c', 'b', 'a', 'b', 'c']
dic = {}
for ch in str1:
   dic.setdefault(ch, 0)
   dic[ch] += 1
   print(list(dic.items()))

print('alphabet counting :', dic)
```

- Se uma chave e valores padrão forem especificados, tais como setdefault(chave, valor), os valores padrão são armazenados nos valores e os valores são devolvidos.

```python
# Exemplo: adiciona uma chave 'f' a um dicionário chamado dic, armazena 50 em um valor padrão e o imprime.
dic = {'a': 10, 'b': 20, 'c': 30, 'd': 40}
print(dic)
dic.setdefault('f', 50)
print(dic)
```

**Método Update (key=value)**
- Modifica o valor da chave no dicionário.
- Por exemplo, se o dicionário for dic = {'a' : 10} (onde a chave é uma string), como mostrado em dic.update (a = 90), os nomes e valores da chave podem ser especificados, exceto citações simples ou duplas da chave.

```python
dic = {'a': 10, 'b': 20, 'c': 30, 'd': 40}
print(dic)
dic.update(a = 90)
print(dic)
```

```python
# Adicionar chave
dic = {'a': 10, 'b': 20, 'c': 30, 'd': 40}
print(dic)
dic.update(e = 50)
print(dic)
```

```python
# Mudar e Adicionar
dic = {'a': 10, 'b': 20, 'c': 30, 'd': 40}
print(dic)
dic.update(a = 900 , f = 60)
print(dic)
```

- Precauções para o uso do método update: A forma de atualização (chave=valor) só pode ser usada quando a chave é uma string.

```python
# Se a chave for um número, o valor pode ser modificado pela inserção de um dicionário como update(dictionary).
# Chaves numericas: use { }
dic2 = {1: 'um', 2: 'dois'}
print(dic2)
dic2.update({1: 'UM', 3: 'TRÊS'}) # não vai dar erro!
# dic2.update(1: 'UM') # vai dar erro!
print(dic2)

dic2 = {1: 'um', 2: 'dois'}
print(dic2)
dic2.update(1: 'UM')
#dic2.update({1: 'UM'}) #vai dar erro!
print(dic2)
```

```python
# Chaves numericas: usando listas
dic2 = {1: 'UM', 2: 'dois', 3: 'TRÊS'}
print(dic2)
dic2.update([[2, 'DOIS'], [4, 'QUATRO']])
print(dic2)
```

```python
# Chaves numericas: usando tuplas
dic2 = {1: 'UM', 2: 'dois', 3: 'TRÊS'}
print(dic2)
dic2.update(((2, 'DOIS'), (4, 'QUATRO')))
print(dic2)
```

**Diferença entre setdefault e update**
- O setdefault só pode adicionar pares de chaves de valor, e os valores das chaves já existentes não podem ser modificados.
- No entanto, o update pode tanto adicionar como modificar o par de valores-chave.
- Mesmo se a chave 'a' já incluída for armazenada como 90 usando setdefault, o valor de 'a' não muda.

```python
dic = {'a': 10,'b':20,'c': 30, 'd':40}
print(dic)
dic.setdefault('a', 90)
print(dic)
dic.update(a = 90)
print(dic)
```

- Cálculo Subset do subconjunto e do superconjunto
  - Se todos os elementos de um conjunto A pertencem a outro conjunto B, então A é um subconjunto de B.
  - B é um superconjunto de A porque B contém tudo o que A tem.

```python
# Método para saber se um conjunto é subconjunto de outro
s1 = {1, 2, 3, 4, 5}
s2 = {1, 2, 3} # é subconjunto do s1
s3 = {1, 2, 6} # não é o subconjunto do s1

print (s2.issubset(s1)) # perguntando se s2 é subconjunto de s1
print (s3.issubset(s1)) # perguntando se s3 é subconjunto de s1
print (s1.issuperset(s2)) # perguntando se s1 é subconjunto de s2
print (s1.issuperset(s3)) # perguntando se s1 é subconjunto de s3
```

```python
# Disjunção entre conjuntos
s1 = {1, 2, 3}
s2 = {10, 20, 30}
s1.isdisjoint(s2) # verificando se s1 e s2 são disjuntos
```

**Função zip e tipos de dados repetíveis**
- Dados repetíveis: listas, dicionários, conjuntos e tuplas
- Quando vários tipos de dados repetíveis são inseridos, a função que retorna o iterador tupla, combinando-os, é a função zip.
- Agregação: o zip pode-se receber vários tipos de dados que podem ser repetidos.

```python
# zip(*itens)
mpty_iterator = zip()
result = set(empty_iterator)
result
```

```python
# Agregação
a = [10, 20, 30] #lista a
b = ('dez', 'vinte', 'trinta') #tupla b
for val in zip(a, b): #print tupla criada pela agregação da lista a e tupla b
    print(val)
```

- Precauções para o uso da função zip

```python
# Quando um fator de tipo de dado repetível é transferido para a função zip, ele aparece como um objeto tupla ('a',), e não como 'a'
lst = ['a', 'b', 'hello', 'this']
my_iterator = zip(lst)
result = set(my_iterator)
result
```

```python
# Agregação: Se houver espaço vazio na coluna, ele é jogador fora por um zip
str_list = [ 'hello', 'world' , 'python', 'rocks']
int_tuple = (1, 2, 3)
int_list = [400, 500, 600, 700, 800]
my_iterator = zip(str_list, int_tuple, int_list)
list(my_iterator)
```

- Descompactação de zip e tupla
  - É a operação de retorno de tuplas vinculadas por meio de uma função zip aos dados repetíveis originais.
  - Você precisa inserir os dados vinculados à função zip como um fator e colocar o operador * na frente do fator.

```python
str_list = [ 'hello', 'world' , 'python', 'rocks']
int_tuple = (1, 2, 3)
int_list = [400, 500, 600, 700, 800]
zip_lst = list(zip(str_list, int_tuple, int_list))
a, b, c = zip(*zip_lst) #descompactar tupla
print(a, b, c)
```
