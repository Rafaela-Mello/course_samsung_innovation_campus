## **Unidade 30: Algoritmo Guloso (Greedy)**

**Abordagem gananciosa**
- É uma técnica de projeto de algoritmo que chega a uma solução selecionando uma "melhor" escolha no momento a partir de uma sequência de escolhas de acordo com critérios pré-definidos.
- Resolve o problema nas três etapas seguintes.
  - Processo de seleção: Escolha o próximo elemento a ser adicionado ao conjunto.
  - Verificação de Viabilidade: Verificar se um novo conjunto é apropriado para ser a resposta.
  - Verificação da solução: Determinar se um novo conjunto é a resposta para o problema.

```python
# Problema do troco
# A entrada do algoritmo é o tipo de moeda e a quantidade de troco, e a saída é uma lista de moedas.

def coin_change (coins, amount):
  changes = []
  largest = 0
  while amount  > 0:
    if amount < coins[largest]:
      largest += 1
    else:
      changes.append(coins[largest])
      amount -= coins[largest]
  return changes

coins = [500, 100, 50, 10]
amount = int(input("Input the amount: "))
changes = coin_change(coins, amount)
print (changes, len(changes))
```

## **Unidade 31: Dividir e Conquistar**

**Dividir e Conquistar**
- A estratégia de dividir e conquistar divide a instância de entrada de um determinado problema em duas ou mais pequenas instâncias de entrada.
- É uma estratégia para resolver os problemas nas seguintes etapas
  - Dividir: Dividir a instância de entrada do problema em duas ou mais pequenas instâncias de entrada.
  - Conquistar: Resolva cada uma das instâncias de entrada divididas. Se a instância dividida não for pequena o suficiente, use a recursividade para resolvê-la.
  - Combinar: Se necessário, encontre a resposta da instância de entrada original, combinando as respostas da pequena instância de entrada.

- Exemplos: Pesquisa binária, quick sort...

## **Unidade 32: Programação Dinâmica**

**Programação Dinâmica X Dividir e Conquistar**
- A programação dinâmica é o mesmo método que o método dividir e conquistar, na medida em que ambos são soluções recursivas para um determinado problema.
- Como o método dividir e conquistar repete cálculos ineficientes quando o subproblema sobreposto ocorre, o método de programação dinâmica armazena os subproblemas em uma variável ou tabela para que os subproblemas já calculados não sejam recalculados.

image.png

**O Projeto de Algoritmos de Programação Dinâmica**

- A programação dinâmica é uma técnica de projeto algorítmico que analisa as características recursivas de um determinado problema e depois obtém uma resposta de forma **bottom-up**.
- Neste momento, a técnica de armazenar as respostas aos subcapítulos sobrepostos na tabela, a fim de resolver o subcapítulo sobreposto, é chamada de **memorização**.
  - Estabelecer uma equação recursiva para calcular a resposta para o caso de entrada do problema
  - A resposta para todo o caso de entrada é obtida por um método bottom-up que resolve primeiro o "bottom-case " de entrada.
- Utiliza a memorização para cálculos de baixo para cima.
- As respostas para os sub-problemas são armazenadas em uma tabela (ou memória).
- Ao encontrar o mesmo subproblema, ele é tirado diretamente da tabela (ou memória) sem recalcular.

**N-ésimo Termo Fibonacci por Recursão**

```python
# A função fib1() é uma implementação da definição recursiva da sequência Fibonacci através de uma função recursiva.
def fib1(n):
    if n == 0 or n == 1:
       return n
    else:
       return fib1(n - 1) + fib1(n - 2)

N = int(input("Entre com o valor de n:"))
print(fib1(N))
```

```python
# A função fib2() é implementada através de uma função recursiva como uma solução de baixo para cima, aplicando um método de programação dinâmico.
F = {0: 0, 1: 1}

def fib2(n):
    if n in F:
        return F[n]
    else:
       F[n] = fib2(n - 1) + fib2(n - 2)
       return F[n]

N = int(input("Entre com o valor de n: "))
print(fib2(N))
```

```python
# Ao chamar a função fib3(), a lista F, que armazena o valor do termo não calculado como -1, deve ser dada como a entrada.
def fib3(F, n):
    if n <= 1:
        return F[n]
    else:
        if F[n] < 0:
            F[n] = fib3(F, n - 1) + fib3(F, n - 2)
        return F[n]

N = int(input("Input a number: "))
F = [0, 1] + [-1] * (N - 1)
print(fib3(F, N))
```

**O Triângulo de Pascal por Recursão**

```python
# Usando as propriedades recursivas do coeficiente binomial, bin(n, k) pode ser calculado da seguinte forma.
def bin1(n, k):
    if k == 0 or n == k:
        return 1
    else:
        return bin1(n - 1, k - 1) + bin1(n - 1, k)
n = int(input("Input the value of n: "))
k = int(input("Input the value of k: "))
print(f"binomial({n}, {k}) is {bin1(n, k)}")
```

```python
# Aplicando uma programação dinâmica, as expressões recursivas podem ser resolvidas de forma bottom-up utilizando loops, como mostrado abaixo.
def bin2(n, k):
    B= [[0] * (n + 1) for i in range (n + 1)] # Changed the inner list size to (n + 1)
    for i in range(n+1):
      for j in range(i+1):
        if j == 0 or j == i: # Changed j == 1 to j == i
          B[i][j] = 1
        else:
            B[i][j] = B[i-1][j-1] + B[i-1][j]

    return B[n][k]

n = int(input("Input the value of n: "))
k = int(input("Input the value of k: "))
print(f"binomial({n}, {k}) is {bin2(n, k)}")
```

## **Unidade 33: Backtracking (Retrocesso)**

**Busca em profundidade**
- A busca de Profundidade refere-se a um método de busca no qual a prioridade de profundidade é pesquisada da seguinte forma.
  - Primeiro, visite o nó raiz, seguido pelo nó 2 e o nó 3.
  - O nó 3 não tem mais nós para visitar, então você tem que devolver o nó anterior...

**Estratégias de backtracking**
- O backtracking é uma estratégia útil de projeto de algoritmos que, incrementalmente, constrói um candidato a solução, e faz o backtracking assim que determina que a solução candidata não pode possivelmente levar a uma solução válida.

- O retrocesso resolve problemas com o seguinte processo
  - Organizar o espaço de busca na forma de uma árvore.
  - Explorar a árvore do espaço de busca por profundidade-primeira busca.
  - Ele retorna quando já não é mais promissor.

- O retrocesso utiliza a técnica de poda da seguinte maneira.
  - Poda refere-se ao retorno sem visitar os nós da sub-árvore que não são promissores.
  - Promissor refere-se a um estado no qual uma solução pode existir na sub-árvore de um nó visitante no momento
