**Parâmetro e Argumento de função**
- **Parâmetro**: É uma **variável** definida no cabeçalho da função ou método. Ela recebe o valor real quando a função é chamada.
  - Exemplo: def foo(m, n): **m** e **n** são **parâmetros**
- **Argumento**: Refere-se ao **valor** real passado quando uma função ou método é chamado. É simplesmente chamado de argumento.
  - Exemplo: foo(3, 4): **3** e **4** são **argumentos** da função foo()

**Parâmetro arbitrário**
- Há casos em que o número de argumentos não é determinado. Isto é chamado de argumento arbitrário.
- Ao receber um argumento coloque um asterisco (*) na frente do parâmetro.
- Os parâmetros arbitrários podem ser usados em declarações for, como tuplas e listas.

```python
# Criar uma função que funcione com 3 ou 2 argumentos através da passagem arbitrária de argumentos
def greet(*nomes):
	for nome in nomes:
		print('Olá ', nome, '!')
greet('A', 'B', 'C') #3 argumentos
greet('James', 'Thomas') #2 argumentos
```

```python
# Contar o número de argumentos passados arbitrariamente usando a função len
def foo(*args):
    print('Número de argumentos: ', len(args))
    print('Argumentos: ', args)
foo(10, 20, 30)
```

```python
# Crie um programa que encontre a soma dos números inseridos como argumentos.
def sum_nums(*numbers):
    result = 0
    for n in numbers:
        result += n
    return result
print(sum_nums(10, 20, 30)) #Imprime a soma dos argumentos 10, 20 e 30
print(sum_nums(10, 20, 30, 40, 50)) #Imprime a soma dos argumentos 10, 20, 30, 40 e 50
```

```python
# Conjunto de produtos: uma função chamando outra função
def product_set(set1, set2):
  res = set()
  for i in set1:
    for j in set2:
      res = res | {(i, j)} #product_set usa um duplo for loop
  return res

def exp(input_set, exponent): # Uma função que eleva input_set a potência exponent
  res = input_set # reinicia
  for _ in range(exponent - 1):
    res = product_set(res, input_set) #uma função chamada dentro de outra função
  return res

A = {1, 3}
A3 = exp(A, 3) # Executa A a potencia 3
print(A3)
```

```python
# Função product_set para contar o número de casos em que um dado é lançado duas vezes.
def product_set(set1, set2):
  res = set()
  for i in set1:
    for j in set2:
      res = res | {(i, j)} #product_set usa um duplo for loop
  return res

cases = { 1, 2, 3, 4, 5, 6 }
cases_2times = product_set(cases, cases)
print(cases_2times)
```

```python
def factorial(n): # Implementar a função recursiva de n!
    if n <= 1:    # Condição de término necessária
 	    return 1
    else:
        return n * factorial(n-1) # n*n(-1)! implementação por definição

n = int(input('Insira um numero: '))
print('{}! = {}'.format(n, factorial(n)))
```

**Sequência de Fibonacci**
- É uma seqüência na qual o primeiro e segundo termos são 1, e todos os termos subsequentes são a soma dos dois termos imediatamente anteriores.
- 0, 1, 1, 2, 3, 5, 8, 13, 21...

```python
def fibonacci(n):
    if n <=1:
        return n
    else:
        return(fibonacci(n-1) + fibonacci(n-2))
nterms = int(input('Quantos números de Fibonacci você quer? '))

if nterms <= 0: # Não é possível encontrar o número de Fibonacci se ele for negativo
    print('Erro: Insira um número positivo')
else:
    print('Sequência de Fibonacci: ', end=' ')
    for i in range(nterms):
        print(fibonacci(i), end=' ')
```

```python
# Implementação da função recursiva da sequência Fibonacci: ver o resultado pelo índice

def fibonacci(n):
    if n <= 1:
        return n
    else:
        return fibonacci(n-1) + fibonacci(n-2) # O valor é agregado como é chamado recursivamente até que a condição de rescisão seja cumprida.
n = int(input('Insira um numero: '))
print(fibonacci(n))
```

- Problema recursivo: não é eficiente, pois repete o cálculo sem sentido, mesmo que seja um valor já calculado.

**Função Fibonacci e memorização**
- A memorização é um método para armazenar os resultados de pequenos problemas que se repetem repetidamente, mas os resultados não mudam.
- Programação Dinâmica (DP): é uma das técnicas de solução de problemas do algoritmo a ser tratada durante o curso do algoritmo.
  - Na programação dinâmica, você tem que resolver pequenos problemas para resolver os grandes problemas.

```python
from timeit import * # importa todas as classes do método timeit
dic = {0:0, 1:1} # Declare um dicionário para memotização como variável global, dic = {0:0, 1:1} Chamando fibonacci_2(1) retorna 1, que é dic[1].

def fibonacci_2(n):
    if n in dic:
        return dic[n] # Se o dicionário já tem um resultado computado, ele retorna o resultado e não faz uma chamada recursiva.
    dic[n] = fibonacci_2(n-1) + fibonacci_2(n-2) # Se o dicionário não tiver resultado, a função faz uma chamada recursiva, e o resultado é armazenado em um dicionário.
    return dic[n]

t2 = Timer('fibonacci_2(30)', 'from __main__ import fibonacci_2')
print('tempo de fibonacci_2(30) * 20: ', t2.timeit(number = 20), ' segundos')
```

## **Unidade 19: Lambda**

**Função lambda**
- É uma função sem nome e também é chamada de expressão lambda. Devido a essa característica, ela também é chamada de função anônima.
- Uma **função regular** consiste na palavra-chave def, nome da função, parâmetros, dois pontos e o corpo, enquanto uma **função lambda** tem apenas as funcionalidades de parâmetro e função.
- Sintaxe: **lambda [parâmetros] : {expressão} : {expressão}**

image.png

```python
# Exemplo: retorna uma soma de números inteiros usando a função adicionar.
def add(x, y):
	return x + y

print('soma de 100 e 200: {}'.format(add(100,200)))
```

```python
# Use uma função lambda para executar o que a função adicionar faz sem definir a função.
add = lambda x, y: x + y
print('soma de 100 e 200:', add(100, 200))
```

```python
# A função Lambda funciona bem mesmo com uma declaração tão simples. Mesmo a variável de adição não é necessária.
print('soma de 100 e 200 :', (lambda x, y: x + y)(100, 200))
```

```python
# A função lambda permite uma codificação simples.
a = [1, 2, 3, 4, 5, 6, 7]
square_a = list(map(lambda x: x**2, a))
print(square_a)
```

**Função filter**
- Recebe elementos iteráveis e retorna somente aqueles que são Verdadeiros em um grupo.
- Sintaxe: **filter((function_to_be_applied, {itarable_object})**

```python
# Exemplo: Há uma lista que armazena as idades de muitas pessoas. Construa um código que extrai as idades de adultos acima de 19 anos.
# Retornar True para valores acima de 19, e False para aqueles que não o são.
def adult_func(n):
    if n >= 19:
        return True
    else:
        return False
ages = [34, 39, 20, 18, 13, 54]
print("lista de adultos:")
for a in filter(adult_func, ages):
    print (a, end = ' ') # filtragem envelhece através da função filter ()
```

```python
# Você pode simplificar a função adult_func transformando-a em uma função anônima usando a palavra-chave lambda.
ages = [34, 39, 20, 18, 13, 54]
print("lista de adultos:")
for a in filter (lambda x: x >= 19, ages): # idade do filtro usando a função de filtro ()
	print (a, end = ' ')
```

**Função map**
- Esta função retorna um datatype iterável aplicando a função de mapeamento em cada unidade do datatype iterável.
- Sintaxe: **map(function_to_be_applied , iterable _object, ...)**
- A função map leva dois ou mais argumentos. O primeiro é uma função de mapeamento e o segundo é um objeto iterável, como uma lista que será inserida na função de mapeamento.

```python
# Valor quadrático computado de todas as unidades da lista
def square(x):
	return x ** 2
a = [1, 2, 3, 4, 5, 6, 7]
# aplicar o valor de retorno da função quadrada a cada termo de um
square_a = list(map(square, a))
print (square_a)
```

**Regra de escopo**
- Estabelece uma regra sobre o escopo de acesso e contexto válido para uma variável
  - L(Local): uma variável local definida dentro de uma função
  - E(Enclosing Function Local): um escopo de uma função que contém outra função (função interna)
  - G(Global): uma variável de um namespace em nível de módulo que não está incluída no escopo da função
  - B(Built-in): uma variável do namespace embutido.

Variáveis:
- Se uma variável é definida dentro de uma função, ela se torna uma variável local.
  - Uma variável local desaparece da memória uma vez terminada a função.
- Se for definida fora de uma função, ela se torna uma variável global em nível de módulo.
  - A palavra-chave global significa que a função utilizará uma variável global fora da função.

```python
x = 10
y = 10	# corresponde a G

def foo():
  x = 20  # função foo corresponde a L
  def bar():
    a = 30	# função  bar corresponde a E
    print(a, x, y)	# cada variável corresponde a E, L e G
  bar()

foo()
x = 40
y = 11
foo()
```

**Closure**
- Cria uma função aninhada dentro de uma função e retorna a função aninhada como um valor de retorno.
- Pode-se dizer que a função aninhada 'foi encerrada'. Se a função for fechada, a variável dentro do corpo da função original (função externa) não desaparece da memória. Então, na próxima chamada, ela pode utilizar a função aninhada.

- Variáveis livres: é usada apenas em um determinado bloco de código, mas não é uma variável global, e não é definida dentro do bloco. É um conceito relativo.

```python
# Se existe uma função externa que encerra a função closure, o tipo de retorno desta função externa se torna o closure.
# Isto é, do ponto de vista da função mult, a é uma variável livre. No entanto, para o closure_calc, a é apenas uma variável local.
def closure_calc(x):
  a = 2
  def mult(x):
    return a * x
  return mult(x)
print(closure_calc(1), closure_calc(2), closure_calc(3))
```

- Principal uso do closure
  - Limitação do uso de variáveis globais
  - Ocultação de dados

```python
# Um closure tem uma memória separada para cada variável.
def makecounter():
  count = 0
  def counter():
    nonlocal count
    count += 1
    return count
  return counter

c1 = makecounter()
c2 = makecounter()

print("c1",c1())
print("c1",c1())
print("c2",c2())
```

```python
# Como gerar um closure
# Gera um closure escrevendo uma função aninhada e devolvendo-a
def calc():
	a = 3
	b = 5
	def mul_add(x):
		return a * x + b
	return mul_add
C = calc()
print(C(1), C(2), C(3), C(4), C(5))
```

```python
# Exemplo de código para fazer closure com lambda: criar uma expressão lambda como return lambda x: a* x + b e devolver a própria expressão lambda.
def calc():
  a = 3
  b = 5
  return lambda x: a * x + b # fazendo fechamento com lambda
C = calc()
print(C(1), C(2), C(3), C(4), C(5))
```

```python
# Exemplo de código de mudança de uma variável local de um closure: usar o nonlocal para mudar a variável local de um closure
def calc():
  a = 2
  b = 3
  total = 0
  def mul_add(x):
    nonlocal total # encontra a variável local mais próxima
    total = total + a*x + b
    return total
  return mul_add

C = calc()
print(C(1), C(2), C(3), C(4), C(5))
```

**Encapsulamento de Classe**

```python
class Cat:
  def __init__(self,name,age):
    self.__name = name
    self.__age = age

  def set_age(self,age):
    if age > 0:
      self.__age = age


  def get_age(self):
    return self.__age

  def __str__(self):
    return('Cat name:'+self.__name+' Age:'+str(self.__age))


espetinho = Cat('Espetinho',3)
print(espetinho)
espetinho.set_age(4)
print(espetinho)
espetinho.set_age(-3)
print(espetinho)
```

```python
list_a = [10, 20, 30]	# list_a que se refere a um objeto de lista
list_b = [10, 20, 30]	# list_b que se refere a um objeto de lista

if list_a is list_b:	# Verifica se os dois objetos da lista são idênticos
	print('list a is list b')
else:
	print('list a is not list b')

print('id(list_a) =',id(list_a),', id(list_b) = ',id(list_b))
```

```python
a = 'ABC'	# Variável a que se refere a um objeto de string
b = 'ABC'	# Variável b que se refere a um objeto de string

if a is b:
    print('a is b')
else:
    print('a is not b')	# Os objetos de string a e b referem-se ao mesmo objeto
```

```python
class Vector2D:
  def __init__(self, x, y):
   self.x = x
   self.y = y

  def __str__(self):
    return "({}, {})".format(self.x, self.y)

  def __add__(self, other):
    return Vector2D(self.x + other.x, self.y + other.y)

  def __sub__(self, other):
    return Vector2D(self.x - other.x, self.y - other.y)

v1 = Vector2D(30, 40)
v2 = Vector2D(10, 20)

v3 = v1.__add__(v2)
print('v1.add(v2)', v3)
v4 = v1.__sub__(v2)
print('v1.add(v2)', v4)
```
