## **Unidade 27: Ordenação por flutuação, seleção e inserção**

**Bubble Sort - Ordenação por Flutuação**
- É um algoritmo de classificação que compara dois elementos adjacentes e troca suas posições se a ordem for invertida
image.png

```python
# Bubble Sort
def bubblesort(S):
  n = len(S)
  for i in range(n):
    print(S)
    for j in range(n - 1):
      if S[j] > S[j + 1]:
        S[j], S[j + 1] = S[j + 1], S[j]

S = [50, 30, 40, 10, 20]
bubblesort(S)
print(S)
```

**Selection Sort - Ordenação por Seleção**: pode usar uma balança de equilíbrio
- É um algoritmo de classificação que encontra o menor elemento em uma lista e o coloca no lado esquerdo. Para isso, ele usa uma estratégia para comparar o primeiro elemento de uma lista não classificada com o resto da lista e para trocar posições com o elemento menor.

image.png



```python
# Selection Sort - Implementação sem troca
def selectionsort1(S):
  R = []
  while len(S) > 0:
    print (R, S)
    smallest = S.index(min(S))
    R.append(S[smallest])
    S.pop(smallest)
  return R

S = [50, 30, 40, 10, 20]
R = selectionsort1(S)
print(R)
```

- A ordenação In-Place refere-se a um algoritmo de ordenação que se classifica através de uma operação de troca sem utilizar memória adicional. O selectionsort1() implementado anteriormente utiliza memória adicional R, portanto não é uma ordenação no local.

```python
# Selection Sort - Implementando In-Place
def selectionsort2(S):
	n = len(S)
	for i in range(n - 1):
		print(S)
		smallest = i
		for j in range(i + 1, n):
			if S[j] < S[smallest]:
				smallest = j
		S[i], S[smallest] = S[smallest], S[i]
S = [50, 30, 40, 10, 20]
selectionsort2(S)
print(S)
```

**Implementação do Merge Sort com memória adicional**

```python
# Fusão de espécies pelo método de dividir e conquistar
# A função merge1() cria uma lista ordenada S pela fusão de L e R, que são listas ordenadas, respectivamente.
def merge1(S, L, R):
  k= 0
  while len(L) > 0 and len(R) > 0:
    if L[0] <= R[0]:
      S[k] = L.pop(0)
    else:
      S[k] = R.pop(0)
    k += 1
  while len(L)!= 0:
    S[k] = L.pop(0)
    k += 1
  while len(R) != 0:
    S[k] = R.pop(0)
    k+= 1

def mergesort1(S):
  n = len(S)
  if n > 1:
    print(S)
    mid = n // 2
    L,R= S[:mid], S[mid:]
    mergesort1(L)
    mergesort1(R)
    merge1(S, L, R)

S = [27, 10, 12, 20, 25, 13, 15, 22]
mergesort1(S)
print(S)
```

```python
# Implementando a Merge Sort de forma mais eficiente
# A mergesort2() classifica S especificando as faixas de baixa e alta ao invés de dividir S em L e R.

def merge2(S, low, mid, high):
    R = []
    i, j = low, mid + 1
    while i <= mid and j <= high:
        if S[i] < S[j]:
            R.append(S[i])  # Correct indentation here
            i += 1
        else:
            R.append(S[j])  # And here
            j += 1
    if i > mid:
        for k in range(j, high + 1):
            R.append(S[k])  # Use S[k] instead of S[j]
    else:
        for k in range(i, mid + 1):
            R.append(S[k])  # Use S[k] instead of S[i]
    for k in range(len(R)):
        S[low + k] = R[k]

def mergesort2(S, low, high):
    if low < high:
        print(S)
        mid = (low + high) // 2  # Assign the result to mid
        mergesort2(S, low, mid)
        mergesort2(S, mid + 1, high)
        merge2(S, low, mid, high)

S = [27, 10, 12, 20, 25, 13, 15, 22]
mergesort2(S, 0, len(S) - 1)
print(S)
```

**Quick Sort** com a abordagem dividir-e-conquistar
- O quick sort utiliza uma estratégia de divisão e conquista semelhante ao merge sort.
- A diferença do merge sort é que, como ela se divide em duas listas baseadas no pivô, os tamanhos das duas listas podem ser diferentes.

```python
def partition1(S, low, high):
  pivot = S[low]
  left, right = low + 1, high
  while left <= right: # Change: Include '=' to handle edge case
    print(S)
    while left <= right and S[left] <= pivot:
      left += 1
    while left <= right and S[right] >= pivot:
      right -= 1
    if left < right:
      S[left], S[right] = S[right], S[left]
  # Change: Ensure pivotpoint is within bounds before swapping
  pivotpoint = min(right, len(S)-1)
  S[low], S[pivotpoint] = S[pivotpoint], S[low]
  return pivotpoint


def quicksort1(S, low, high):
  if low < high:
    print(S)
    pivotpoint = partition1(S, low, high)
    quicksort1(S, low, pivotpoint -1)
    quicksort1(S, pivotpoint + 1, high) # Uncommented this line to complete the quicksort

S = [15, 10, 12, 20, 25, 13, 22]
quicksort1(S, 0, len(S)-1)
print(S)
```
